```python
def cleanNaN(alldatalist,columnIndex=4): #columnIndex = 4 is Total Expenditure 
    counter = 0
    for item in alldatalist: #len = 159 
        # drop nan
        item.dropna(subset=["State"], inplace=True)
        
        counter += 1
        #print(counter)
        
    return alldatalist
```

```python
def cleanString(alldatalist, columnIndex=4):
    counter = 0
    for item in alldatalist:
        item.iloc[:,columnIndex] = item.iloc[:,columnIndex].apply(
            lambda gstr: float(gstr.replace(",","").replace("$","")))
        counter = counter+1
        #print(counter)
    return alldatalist
```

```python
def plotHist(dfSeries):
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(9,4))
    dfSeries.hist(ax=ax, bins=15, color="steelblue", align='mid',normed=False, edgecolor="white")
    ax.grid(alpha=0.25)
    ax.set_axisbelow(True)
    filename = "/Users/kair/data_mining/DM_VA_DATA/hist{0:02d}"
    counter = 1
    plt.savefig(filename.format(counter))
    plt.xlabel('Number of Veterans')
    plt.ylabel('Frequency of Occurence')
    plt.title('Histogram Showing Number of Veterans by State')
```

```python
def plot2Hist(dfSeries):
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(9,4))
    dfSeries.hist(ax=ax, bins=15, color="steelblue", align='mid',normed=False, edgecolor="white")
    ax.grid(alpha=0.25)
    ax.set_axisbelow(True)
    filename = "/Users/kair/data_mining/DM_VA_DATA/hist{0:02d}"
    counter = 2
    plt.savefig(filename.format(counter))
    plt.xlabel('Ratio of Female veterans to Total Veterans')
    plt.ylabel('Frequency of Occurence')
    plt.title('Histogram Showing Ratio of Female Veterans by State')
```
