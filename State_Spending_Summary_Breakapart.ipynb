```python
#read from Puerto Rico file
#Look at dollars spent on a benefit (y-axis) and number of veterans (x-axis)

import math
import csv

comp_outliers=[]
med_outliers=[]

FY_14_FILE_NAMES=['fy14_ak.csv','fy14_al.csv','fy14_ar.csv','fy14_az.csv','fy14_ca.csv','fy14_co.csv','fy14_ct.csv',
                  'fy14_dc.csv','fy14_de.csv','fy14_fl.csv','fy14_ga.csv','fy14_gu.csv','fy14_hi.csv','fy14_ia.csv',
                  'fy14_id.csv','fy14_il.csv','fy14_in.csv','fy14_ks.csv','fy14_ky.csv','fy14_la.csv','fy14_ma.csv',
                  'fy14_md.csv','fy14_me.csv','fy14_mi.csv','fy14_mn.csv','fy14_mo.csv','fy14_ms.csv','fy14_mt.csv',
                  'fy14_nc.csv','fy14_nd.csv','fy14_ne.csv','fy14_nh.csv','fy14_nj.csv','fy14_nm.csv','fy14_nv.csv',
                  'fy14_ny.csv','fy14_oh.csv','fy14_ok.csv','fy14_or.csv','fy14_pa.csv','fy14_pr.csv','fy14_ri.csv',
                  'fy14_sc.csv','fy14_sd.csv','fy14_tn.csv','fy14_tx.csv','fy14_ut.csv','fy14_va.csv','fy14_vt.csv',
                  'fy14_wa.csv','fy14_wi.csv','fy14_wv.csv','fy14_wy.csv']
FY_15_FILE_NAMES=['fy15_ak.csv','fy15_al.csv','fy15_ar.csv','fy15_az.csv','fy15_ca.csv','fy15_co.csv','fy15_ct.csv',
                  'fy15_dc.csv','fy15_de.csv','fy15_fl.csv','fy15_ga.csv','fy15_gu.csv','fy15_hi.csv','fy15_ia.csv',
                  'fy15_id.csv','fy15_il.csv','fy15_in.csv','fy15_ks.csv','fy15_ky.csv','fy15_la.csv','fy15_ma.csv',
                  'fy15_md.csv','fy15_me.csv','fy15_mi.csv','fy15_mn.csv','fy15_mo.csv','fy15_ms.csv','fy15_mt.csv',
                  'fy15_nc.csv','fy15_nd.csv','fy15_ne.csv','fy15_nh.csv','fy15_nj.csv','fy15_nm.csv','fy15_nv.csv',
                  'fy15_ny.csv','fy15_oh.csv','fy15_ok.csv','fy15_or.csv','fy15_pa.csv','fy15_pr.csv','fy15_ri.csv',
                  'fy15_sc.csv','fy15_sd.csv','fy15_tn.csv','fy15_tx.csv','fy15_ut.csv','fy15_va.csv','fy15_vt.csv',
                  'fy15_wa.csv','fy15_wi.csv','fy15_wv.csv','fy15_wy.csv']
FY_16_FILE_NAMES=['fy16_ak.csv','fy16_al.csv','fy16_ar.csv','fy16_az.csv','fy16_ca.csv','fy16_co.csv','fy16_ct.csv',
                  'fy16_dc.csv','fy16_de.csv','fy16_fl.csv','fy16_ga.csv','fy16_gu.csv','fy16_hi.csv','fy16_ia.csv',
                  'fy16_id.csv','fy16_il.csv','fy16_in.csv','fy16_ks.csv','fy16_ky.csv','fy16_la.csv','fy16_ma.csv',
                  'fy16_md.csv','fy16_me.csv','fy16_mi.csv','fy16_mn.csv','fy16_mo.csv','fy16_ms.csv','fy16_mt.csv',
                  'fy16_nc.csv','fy16_nd.csv','fy16_ne.csv','fy16_nh.csv','fy16_nj.csv','fy16_nm.csv','fy16_nv.csv',
                  'fy16_ny.csv','fy16_oh.csv','fy16_ok.csv','fy16_or.csv','fy16_pa.csv','fy16_pr.csv','fy16_ri.csv',
                  'fy16_sc.csv','fy16_sd.csv','fy16_tn.csv','fy16_tx.csv','fy16_ut.csv','fy16_va.csv','fy16_vt.csv',
                  'fy16_wa.csv','fy16_wi.csv','fy16_wv.csv','fy16_wy.csv']
def isOutlier(fname, index, mean, standard_deviation):
    global comp_outliers
    global med_outliers
    with open(fname, 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            if '(Totals)' in str(row[2]):
                value=row[index].replace('$','')
                value=value.replace(',','')
                value=value.replace(' - ', '0')
                if (float(value)<=(mean-standard_deviation)) or (float(value)>=(mean+standard_deviation)):
                    print("Index==", str(index), ". Value==", str(value))
                    if index==5:
                        comp_outliers.append(fname)
                    else:
                        med_outliers.append(fname)
                return

def standardDeviationSum(fname, index, mean):
    with open(fname, 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            if '(Totals)' in str(row[2]):
                temp=row[index].replace('$','')
                temp=temp.replace(',','')
                temp=temp.replace(' - ', '0')
                value=(float(temp)-mean)**2
                return(value)

def gatherInfo(fname, index):
    with open(fname, 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            if'(Totals)' in str(row[2]):
                #is the totals row for the state
                value=row[index].replace('$', '')
                value=value.replace(',','')
                value=value.replace(' - ', '0')
                return(float(value))

if __name__=="__main__":
    global FY_14_FILE_NAMES
    global FY_15_FILE_NAMES
    global FY_16_FILE_NAMES
    
    compensation_sum14=0
    compensation_sum15=0
    compensation_sum16=0
    
    medical_sum14=0
    medical_sum15=0
    medical_sum16=0
    
    #GATHER COMPENSATION AND MEDICAL SUMS FOR MEAN FOR EACH YEAR
    for file in FY_14_FILE_NAMES: #gather compensation and medical info for FY14
        compensation_sum14=compensation_sum14+gatherInfo(file, 5)
        medical_sum14=medical_sum14+gatherInfo(file, 11)
        
    for file in FY_15_FILE_NAMES: #gather comp and med for FY15
        compensation_sum15=compensation_sum15+gatherInfo(file, 5)
        medical_sum15=medical_sum15+gatherInfo(file, 11)
        
    for file in FY_16_FILE_NAMES:
        compensation_sum16=compensation_sum16+gatherInfo(file, 5)
        medical_sum16=medical_sum16+gatherInfo(file, 11)
        
    compensation_mean14=float(compensation_sum14)/53
    compensation_mean15=float(compensation_sum15)/53
    compensation_mean16=float(compensation_sum16)/53
    
    medical_mean14=float(medical_sum14)/53
    medical_mean15=float(medical_sum15)/53
    medical_mean16=float(medical_sum16)/53
    
    #USE THE MEANS CALCULATED ABOVE TO FIND THE STANDARD DEVIATION
    std_sum_comp14=0
    std_sum_comp15=0
    std_sum_comp16=0
    
    std_sum_med14=0
    std_sum_med15=0
    std_sum_med16=0
    
    print("Means found")
    
    #FIRST FIND THE NUMERATOR (the sum of (value-mean)^2)
    for file in FY_14_FILE_NAMES:
        std_sum_comp14=std_sum_comp14+standardDeviationSum(file, 5, compensation_mean14)
        std_sum_med14=std_sum_med14+standardDeviationSum(file, 11, medical_mean14)
        
    for file in FY_15_FILE_NAMES:
        std_sum_comp15=std_sum_comp15+standardDeviationSum(file, 5, compensation_mean15)
        std_sum_med15=std_sum_med15+standardDeviationSum(file, 11, medical_mean15)
    
    for file in FY_16_FILE_NAMES:
        std_sum_comp16=std_sum_comp16+standardDeviationSum(file, 5, compensation_mean16)
        std_sum_med16=std_sum_med16+standardDeviationSum(file, 11, medical_mean16)
        
    #now divide by N-1
    #compensation
    comp_fraction14=float(std_sum_comp14)/52
    comp_fraction15=float(std_sum_comp15)/52
    comp_fraction16=float(std_sum_comp16)/52
    
    #medical
    med_fraction14=float(std_sum_med14)/52
    med_fraction15=float(std_sum_med15)/52
    med_fraction16=float(std_sum_med16)/52
    
    print("Found Fractions")
    
    #now obtain square root --> Have the standard deviation for each year for each benefit
    std_dev_comp14=math.sqrt(comp_fraction14)
    std_dev_comp15=math.sqrt(comp_fraction15)
    std_dev_comp16=math.sqrt(comp_fraction16)
    
    std_dev_med14=math.sqrt(med_fraction14)
    std_dev_med15=math.sqrt(med_fraction15)
    std_dev_med16=math.sqrt(med_fraction16)
    
    print("Found Standard Deviations")
    
    #determine the outliers
    print("Mean-Standard Dev Comp FY14-- ", str(compensation_mean14-std_dev_comp14))
    print("Mean-Standard Dev Med FY14-- ", str(medical_mean14-std_dev_med14))
    print("Mean+Standard Dev Comp FY14-- ", str(compensation_mean14+std_dev_comp14))
    print("Mean+Standard Dev Med FY14-- ", str(medical_mean14+std_dev_med14))
    for file in FY_14_FILE_NAMES:
        isOutlier(file, 5, compensation_mean14, std_dev_comp14)
        isOutlier(file, 11, medical_mean14, std_dev_med14)
        
    print("Mean-Standard Dev Comp FY15-- ", str(compensation_mean15-std_dev_comp15))
    print("Mean-Standard Dev Med FY15-- ", str(medical_mean15-std_dev_med15))
    print("Mean+Standard Dev Comp FY15-- ", str(compensation_mean15+std_dev_comp15))
    print("Mean+Standard Dev Med FY15-- ", str(medical_mean15+std_dev_med15))
    for file in FY_15_FILE_NAMES:
        isOutlier(file, 5, compensation_mean15, std_dev_comp15)
        isOutlier(file, 11, medical_mean15, std_dev_comp15)
    
    print("Mean-Standard Dev Comp FY16-- ", str(compensation_mean16-std_dev_comp16))
    print("Mean-Standard Dev Med FY16-- ", str(medical_mean16-std_dev_med16))
    print("Mean+Standard Dev Comp FY16-- ", str(compensation_mean16+std_dev_comp16))
    print("Mean+Standard Dev Med FY16-- ", str(medical_mean16+std_dev_med16))
    for file in FY_16_FILE_NAMES:
        isOutlier(file, 5, compensation_mean16, std_dev_comp16)
        isOutlier(file, 11, medical_mean16, std_dev_comp16)
        
    print("Found Outliers")
    print("The outliers are:")
    print("Compensation Outliers -- ", str(comp_outliers))
    print("FY14 Compensation Mean -- ", str(compensation_mean14))
    print("FY15 Compensation Mean -- ", str(compensation_mean15))
    print("FY16 Compensation Mean -- ", str(compensation_mean16))
    print("Medical Outliers -- ",med_outliers)
    print("FY14 Medical Mean -- ", str(medical_mean14))
    print("FY15 Medical Mean -- ", str(medical_mean15))
    print("FY16 Medical Mean -- ", str(medical_mean16))
        
    
```
